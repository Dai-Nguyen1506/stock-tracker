## 2. Một vài sàn giao dịch hỗ trợ

```python
import requests

API_KEY = Config.APCA_API_KEY_ID
API_SECRET = Config.APCA_API_SECRET_KEY

url = "https://paper-api.alpaca.markets/v2/assets"
headers = {
    "accept": "application/json",
    "APCA-API-KEY-ID": API_KEY,
    "APCA-API-SECRET-KEY": API_SECRET
}
params = {
    "asset_class": "crypto", # Chỉ lấy danh sách Crypto
    "status": "active"       # Chỉ lấy các mã đang hoạt động
}

response = requests.get(url, headers=headers, params=params)
crypto_assets = response.json()

# Lọc ra danh sách các Symbol
alpaca_crypto_symbols = [asset['symbol'] for asset in crypto_assets]

print(f"Alpaca hiện đang hỗ trợ {len(alpaca_crypto_symbols)} mã Crypto.")
print(alpaca_crypto_symbols) # In thử 20 mã đầu tiên
```
